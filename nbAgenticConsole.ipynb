```python
from rich.console import Console
from rich.panel import Panel
from rich.text import Text

from agentic_console import build_orchestrator
from config_agent import print_wrap

def print_rich_reflection(data: dict):
    console = Console()
    console.print(Panel.fit(
            Text(data.get('summary', ''), style="cyan"),
            title=f"🤖 Agent Reflection — {'✅ SUCCESS' if data.get('success') else '❌ FAILED'} "
                f"({data.get('steps_ok')}/{data.get('steps_total')} steps)",
            border_style="green" if data.get('success') else "red"
        )
    )

objOrchestrator = build_orchestrator(allow_writes=True)
```

# Example 1

```python
# Example goals — the TPA plans, the TSA fills in arguments, and the
# action layer executes each step under sandboxed permissions.
dictReflection = objOrchestrator.run(
    "Update NAVs for owner SG, record today's history, then give me "
    "a performance review and flag any fund down more than 8% from its peak.",
    owner_name="SG",
)

print_wrap("\n" + "=" * 80 + "\n")

print_rich_reflection(dictReflection)
```
