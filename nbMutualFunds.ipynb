## <font color='blue'>Fetch and Update Lowest and Highest NAV</font>

```python
listFunds = objConsole.UniqueFundNames()
print(len(listFunds), textwrap.fill(str(listFunds), 80))
```

```python
for nI, fund_name in enumerate(listFunds):
    dictNAV_Data = objConsole.GetFundHigestNAV(fund_name)
    #print(textwrap.fill(str(dictNAV_Data), width=80))
    nav_highest = dictNAV_Data['nav_highest']
    print('#', nI, nav_highest)
    objConsole.ReviseHighestNav(fund_name, nav_highest)

print('DONE')
```

```python
for nI, fund_name in enumerate(listFunds):
    dictNAV_Data = objConsole.GetFundLowestNAV(fund_name)
    #print(textwrap.fill(str(dictNAV_Data), width=80))
    nav_lowest = dictNAV_Data['nav_lowest']
    print('#', nI, nav_lowest)
    objConsole.ReviseLowestNav(fund_name, nav_lowest)

print('DONE')
```
